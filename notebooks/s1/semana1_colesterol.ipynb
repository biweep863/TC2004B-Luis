### Imports

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.tools.tools import add_constant
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
df_auto = pd.read_csv("../data/s1/e4/auto-mpg.csv")
print(df_auto.head())
```

```python
# Tipos de datos y verificación de horsepower
print("\nTipos de datos:")
print(df_auto.dtypes)
```

```python
# Valores faltantes por columna
print("\nValores faltantes:")
print(df_auto.isnull().sum())
```

```python
# Identificar valores no numéricos en horsepower
hp_str = df_auto["horsepower"].astype(str).str.strip()
non_numeric_hp = hp_str[~hp_str.str.match(r"^\d+(\.\d+)?$")].unique()
print("\nValores no numéricos en horsepower:")
print(non_numeric_hp)
```

```python
# Reemplazar inválidos por NaN y convertir a numérico
df_auto["horsepower"] = pd.to_numeric(
    df_auto["horsepower"].replace(non_numeric_hp, pd.NA),
    errors="coerce"
)
print("\nValores faltantes después de convertir horsepower:")
print(df_auto.isnull().sum())
```

```python
# Rellenar NaN en horsepower con la media
hp_mean = df_auto["horsepower"].mean()
df_auto["horsepower"] = df_auto["horsepower"].fillna(hp_mean)
print("\nValores faltantes después de fillna:")
print(df_auto.isnull().sum())
```

```python
# Verificar tipos de datos
print("\nTipos de datos después de limpiar:")
print(df_auto.dtypes)
```

```python
# Verificar tipos de datos
print("\nTipos de datos después de limpiar:")
print(df_auto.dtypes)
```

```python
# Correlación y heatmap
corr = df_auto.corr(numeric_only=True)
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Pairplot de variables numéricas
sns.pairplot(df_auto.select_dtypes(include=[np.number]))
plt.show()
```

```python
# VIF
df_vif = df_auto.drop(columns=["car name", "origin"], errors="ignore")
df_vif = df_vif.select_dtypes(include=[np.number]).dropna()
X = add_constant(df_vif)
vif_data = pd.DataFrame({
    "feature": X.columns,
    "VIF": [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
})
print("\nVIF:")
print(vif_data)
```
